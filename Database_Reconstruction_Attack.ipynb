```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

from art.attacks.inference.reconstruction import DatabaseReconstruction

import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
```

# Task 1: Logistic regression on Iris dataset

## Training Model

```python
dataset = datasets.load_iris()

# Convert the dataset to a Pandas DataFrame
df = pd.DataFrame(data=np.c_[dataset['data'], dataset['target']], columns=dataset['feature_names'] + ['target'])

#Displaying the dataset
df
```

```python
X=dataset['data']
y=dataset['target']

#Scaling the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=37)
```

```python
# Initialize model
iris_reg = LogisticRegression(max_iter=5000)
```

```python
# Training the Logistic regression model
iris_reg.fit(X_train, y_train)
```

```python
# Predict the labels for the test set
y_pred = iris_reg.predict(X_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy of Logistic Regression on Iris dataset:", accuracy)
```

## Reconstruction

```python
from art.estimators.classification.scikitlearn import ScikitlearnLogisticRegression

#Creating an ART estimator for logistic regression
art_iris_reg = ScikitlearnLogisticRegression(iris_reg)
```

```python
#Defining a database reconstruction instance based on Sci Kit learn class
dbrecon = DatabaseReconstruction(art_iris_reg)
```

```python
num_sample_recon1=0
recon1_state=[]
recon1_RMSE=[]

for i in range(X_train.shape[0]):
  print("i= ",i)
  victim=i
  X_train_adversary=np.delete(X_train,victim,axis=0)
  y_train_adversary=np.delete(y_train,victim,axis=0)

  x, y = dbrecon.reconstruct(X_train_adversary, y_train_adversary)

  if(np.argmax(y) == y_train[victim]):
    num_sample_recon1+=1
    print("Success")
    recon1_state.append(1)
  else:
    print("Failed")
    recon1_state.append(0)

  RMSE=np.sqrt(((X_train[victim] - x) ** 2).sum() / X_train.shape[1])
  recon1_RMSE.append(RMSE)
  print("Inference RMSE: ",RMSE)
```

```python
#For the 2 class classification problem we will be using the
# Breast cancer Wisconcin (Diagnostic) dataset
# which is a binary classification problem

from sklearn.datasets import load_breast_cancer

#Load breast cancer dataset
cancer_dset=load_breast_cancer()

# Create a DataFrame from the dataset's data and target attributes
df2=pd.DataFrame(data=cancer_dset.data, columns=cancer_dset.feature_names)

# Add target labels to the DataFrame
df2['target'] = cancer_dset.target

df2
```

```python
# Initialize model
cancer_reg = LogisticRegression(max_iter=5000)
```

```python
# Training the Logistic regression model
cancer_reg.fit(X2_train, y2_train)
```

```python
# Predict the labels for the test set
y2_pred = cancer_reg.predict(X2_test)

# Calculate the accuracy of the model
accuracy2 = accuracy_score(y2_test, y2_pred)
print("Accuracy of Logistic Regression on cancer dataset:", accuracy2)
```

## Reconstruction attack

```python
#We first generate a random array with no repeated entries to randomly select a victim data point in the dataset
#Due to computational constraints we will run this for 50 different victims

import random
# Generate a random array of 50 entries without repetition in the range of 1 to 564
size=X2_train.shape[0]
print(size)
victim_array = random.sample(range(0, size), 10)

print(victim_array)
```

```python
#Creating an ART estimator for logistic regression
art_cancer_reg = ScikitlearnLogisticRegression(cancer_reg)

#Defining a database reconstruction instance based on Sci Kit learn class
dbrecon2 = DatabaseReconstruction(art_cancer_reg)
```

```python
num_sample_recon_2=0
recon2_state=[]
recon_RMSE=[]

for a in victim_array :
  print("victim Id= ",a)
  victim=a
  X_train_adversary=np.delete(X2_train,victim,axis=0)
  y_train_adversary=np.delete(y2_train,victim,axis=0)

  x, y = dbrecon2.reconstruct(X_train_adversary, y_train_adversary)

  if(np.argmax(y) == y2_train[victim]):
    num_sample_recon_2+=1
    recon2_state.append(1)
    print("Success")
  else:
    recon2_state.append(0)
    print("Failed")

  RMSE=np.sqrt(((X2_train[victim] - x) ** 2).sum() / X2_train.shape[1])
  recon_RMSE.append(RMSE)
  print("Inference RMSE: ",RMSE)
```

```python
print("Total number of samples tested for reconstruction: ", 10)
print("Total number reconstructed success  : ",num_sample_recon_2)
print("Success rate: ", (num_sample_recon_2/10)*100, " %")
print("The average reconstruction accuraccy is: ", sum(recon_RMSE)/len(recon_RMSE))
```

2. However the prediction success rate is good. This indicates that the reconstructed samples still fall in the ranges of the respective classes. This indicates an acceptable fit of the model
3. In most situations the reconstructed data points and prediction on the same matches the original victim. However in the cases it doesnt, its possible our reconstructed data point is not accurate.
